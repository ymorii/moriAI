```python
import graphviz
from IPython.display import display

# Function to create Use Case Diagram
def create_use_case_diagram():
    use_case = graphviz.Digraph('UseCaseDiagram', format='png')
    use_case.node('A', 'Admin', shape='ellipse')
    use_case.node('P', 'Pegawai', shape='ellipse')

    use_case.node('UC1', 'Kelola Pengguna', shape='box')
    use_case.node('UC2', 'Kelola Data Barang', shape='box')
    use_case.node('UC3', 'Kelola Stok Barang', shape='box')
    use_case.node('UC4', 'Menerima Notifikasi Stok', shape='box')
    use_case.node('UC5', 'Buat Laporan', shape='box')

    use_case.node('UC6', 'Catat Barang Masuk', shape='box')
    use_case.node('UC7', 'Catat Barang Keluar', shape='box')
    use_case.node('UC8', 'Cek Stok Barang', shape='box')

    use_case.edge('A', 'UC1')
    use_case.edge('A', 'UC2')
    use_case.edge('A', 'UC3')
    use_case.edge('A', 'UC4')
    use_case.edge('A', 'UC5')

    use_case.edge('P', 'UC6')
    use_case.edge('P', 'UC7')
    use_case.edge('P', 'UC8')

    display(use_case)
    print("Use Case Diagram berhasil dibuat!")

# Function to create Activity Diagram
def create_activity_diagram():
    activity = graphviz.Digraph('ActivityDiagram', format='png')
    activity.node('Start', 'Mulai', shape='ellipse', style='filled', fillcolor='lightgreen')
    activity.node('Login', 'Login ke Sistem', shape='box')
    activity.node('CekAkses', 'Cek Hak Akses', shape='diamond')
    activity.node('AdminPanel', 'Akses Panel Admin', shape='box')
    activity.node('PegawaiPanel', 'Akses Panel Pegawai', shape='box')
    activity.node('KelolaBarang', 'Kelola Data Barang', shape='box')
    activity.node('KelolaStok', 'Kelola Stok Barang', shape='box')
    activity.node('BuatLaporan', 'Buat Laporan', shape='box')
    activity.node('CatatMasuk', 'Catat Barang Masuk', shape='box')
    activity.node('CatatKeluar', 'Catat Barang Keluar', shape='box')
    activity.node('CekStok', 'Cek Stok Barang', shape='box')
    activity.node('End', 'Selesai', shape='ellipse', style='filled', fillcolor='lightblue')

    activity.edge('Start', 'Login')
    activity.edge('Login', 'CekAkses')
    activity.edge('CekAkses', 'AdminPanel', label='Admin')
    activity.edge('CekAkses', 'PegawaiPanel', label='Pegawai')
    activity.edge('AdminPanel', 'KelolaBarang')
    activity.edge('AdminPanel', 'KelolaStok')
    activity.edge('AdminPanel', 'BuatLaporan')
    activity.edge('KelolaBarang', 'End')
    activity.edge('KelolaStok', 'End')
    activity.edge('BuatLaporan', 'End')
    activity.edge('PegawaiPanel', 'CatatMasuk')
    activity.edge('PegawaiPanel', 'CatatKeluar')
    activity.edge('PegawaiPanel', 'CekStok')
    activity.edge('CatatMasuk', 'End')
    activity.edge('CatatKeluar', 'End')
    activity.edge('CekStok', 'End')

    display(activity)
    print("Activity Diagram berhasil dibuat!")

# Function to create Swimlane Diagram
def create_swimlane_diagram():
    swimlane = graphviz.Digraph('SwimlaneDiagram', format='png', graph_attr={'rankdir': 'LR'})

    with swimlane.subgraph(name='cluster_admin') as admin:
        admin.attr(label='Admin (Pemilik Warung)')
        admin.node('Login', 'Login ke Sistem', shape='box')
        admin.node('KelolaBarang', 'Kelola Data Barang', shape='box')
        admin.node('KelolaStok', 'Kelola Stok Barang', shape='box')
        admin.node('Notifikasi', 'Cek Notifikasi Stok', shape='box')
        admin.node('BuatLaporan', 'Buat Laporan', shape='box')

    with swimlane.subgraph(name='cluster_pegawai') as pegawai:
        pegawai.attr(label='Pegawai Warung')
        pegawai.node('CatatMasuk', 'Catat Barang Masuk', shape='box')
        pegawai.node('CatatKeluar', 'Catat Barang Keluar', shape='box')
        pegawai.node('CekStok', 'Cek Stok Barang', shape='box')

    swimlane.edge('Login', 'KelolaBarang')
    swimlane.edge('Login', 'KelolaStok')
    swimlane.edge('KelolaStok', 'Notifikasi')
    swimlane.edge('KelolaStok', 'BuatLaporan')
    swimlane.edge('CatatMasuk', 'CekStok')
    swimlane.edge('CatatKeluar', 'CekStok')

    display(swimlane)
    print("Swimlane Diagram berhasil dibuat!")

# Jalankan fungsi untuk membuat diagram
create_use_case_diagram()
create_activity_diagram()
create_swimlane_diagram()

print("Semua diagram telah dibuat!")
```
